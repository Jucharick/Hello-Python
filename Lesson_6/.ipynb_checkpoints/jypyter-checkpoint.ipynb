```python
f = lambda x: random.randint(1,10)
points = list(map(f, range(1,10)))
points
```

```python
x = list(map(f, range(1,10)))
y = list(map(f, range(1,10)))
print(x)
print(y)
matplotlib.pyplot.plot(x, y, 'ro')
matplotlib.pyplot.show()
```

```python
x = list(range(1,200))
fx = [i*i for i in x]
matplotlib.pyplot.axes(facecolor='black')
matplotlib.pyplot.plot(x, fx)
matplotlib.pyplot.show()
```

```python
count = 100 # кубик бросаем count раз
x = list(range(1,7))
y = [0 for i in x]

fig, ax = matplotlib.pyplot.subplots()

for _ in range(count):
    y[random.randint(0,5)]+=1
    
ax.bar(x,y)

y.sort()

stat = [f'{i+1}: {round(y[i] / count * 100)}%' for i in range(0,6)]
print(stat)
matplotlib.pyplot.show()
```
